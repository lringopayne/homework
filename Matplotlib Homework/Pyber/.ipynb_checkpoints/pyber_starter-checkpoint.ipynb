```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset and Display the data table for preview
pyber_data = pd.merge(ride_df, city_df, on="city", how="left")
pyber_data.head(5)

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types - Filter dataframe
#Urban Type
urban_data = pyber_data[pyber_data['type'] == "Urban"]
#Rural Type
rural_data = pyber_data[pyber_data['type'] == "Rural"]
#Surburban Type
suburban_data = pyber_data[pyber_data['type'] == "Suburban"]

```

```python
#Urban Type for Count Per City
urban_count = urban_data.groupby(["city"]).count()["ride_id"]
#Rural Type for Count Per City
rural_count = rural_data.groupby(["city"]).count()["ride_id"]
#Surburban Type for Count Per City
suburban_count = suburban_data.groupby(["city"]).count()["ride_id"]
```

```python
# Average Fare for Urban Type
urban_fare = urban_data.groupby(["city"]).mean()["fare"]
# Average Fare for Rural Type
rural_fare = rural_data.groupby(["city"]).mean()["fare"]
# Average Fare for Surburban Type
suburban_fare = suburban_data.groupby(["city"]).mean()["fare"]

```

```python
# drivercount for bubble chart - Urban
urban_drivercount = urban_data.groupby(["city"]).first()["driver_count"]
# drivercount for bubble chart - Rural
rural_drivercount = rural_data.groupby(["city"]).first()["driver_count"]
# drivercount for bubble chart - Suburban
suburban_drivercount = suburban_data.groupby(["city"]).first()["driver_count"]
```

```python
# Bubble Scatter Plot
# Scatter plot for Urban
Urban = plt.scatter(x = urban_count,y = urban_fare, s= urban_drivercount*10, marker="o", facecolors="red", edgecolors="black", alpha =0.5)
# Scatter plot for Rural 
Rural = plt.scatter(x = rural_count,y = rural_fare, s= rural_drivercount*10, marker="o", facecolors="green", edgecolors="black", alpha =0.5)
# Scatter plot for Suburban
Suburban = plt.scatter(x = suburban_count,y = suburban_fare, s= suburban_drivercount*10, marker="o", facecolors="yellow", edgecolors="black", alpha =0.5)
#labels
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
#Legend
lgnd = plt.legend((Urban, Rural, Suburban),('Urban', 'Rural', 'Suburban'), title="City Type")
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]

#textbox
plt.figtext(1.3, .5,"Note: Circle size correlates with driver count per city", wrap=True,
           horizontalalignment='center', fontsize=12)
#grid
plt.grid()
```

```python
# Save an image of the chart and print to screen
# NOTE: If your plot shrinks after saving an image,
# update matplotlib to 2.2 or higher,
# or simply run the above cells again.
plt.savefig("../Images/Pyberdata.png")
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
#Totals by city type
#Urban
uftotal = urban_data.sum()["fare"]
#Rural
rftotal = rural_data.sum()["fare"]
#Suburban
sftotal = suburban_data.sum()["fare"]
```

```python
# Build Pie Chart
citytypes = ["Suburban", "Rural", "Urban"]
faresize = [sftotal, rftotal, uftotal]
colors = ["yellowgreen", "red", "lightcoral"]
explode = (0.1, 0, 0)

plt.title("% of Total Fares by City Type")
plt.pie(faresize, explode=explode, labels=citytypes, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
#Save Figure
plt.savefig("../Images/farecitytype.png")
#Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
#Urban
urtotal = urban_data.count()["type"]
#Rural
rrtotal = rural_data.count()["type"]
#Suburban
srtotal = suburban_data.count()["type"]
```

```python
# Build Pie Chart
citytypes = ["Suburban", "Rural", "Urban"]
ridesize = [srtotal, rrtotal, urtotal]
colors = ["yellowgreen", "red", "lightcoral"]
explode = (0.1, 0, 0)

plt.title("% of Total Rides by City Type")
plt.pie(ridesize, explode=explode, labels=citytypes, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
#Save Figure
plt.savefig("../Images/ridecitytype.png")
#Show Figure
plt.show()
```
