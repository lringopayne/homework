```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#Creating Data Frame to retrieve specific data
player_data = purchase_data.loc[:,['SN','Age','Gender','Price']]
player_data.head(5)
```

## Player Count

* Display the total number of players


```python
#Total Number of Players
number_players = player_data["SN"].nunique()
number_players

#summary table
summary_table = pd.DataFrame({"Player Count":[number_players]})
summary_table.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#number of unique items
unique_items =len(purchase_data["Item ID"].unique())
#Average Price
avg_price = '${:.2f}'.format(purchase_data["Price"].mean())
#Total Revenue
total_price = '${:.2f}'.format(purchase_data["Price"].sum())

#Total Number of Purchases
purchase_total = len(purchase_data["Price"])

#summary table
summary_table = pd.DataFrame({"Unique Items": unique_items,
                             "Average Price": [avg_price],
                             "Total Purchases": purchase_total,
                             "Total Revenue": total_price})
summary_table.head()
```

## Age Demographics

* Establish bins for ages

* Calculate the numbers and percentages by age group

* Optional: round the percentage column to two decimal points

```python
# Establish bins for ages

bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


player_data["Age Summary"] = pd.cut(player_data["Age"], bins, labels=group_names)

#Total Count by age
demographic_data_counts = player_data["Age Summary"].value_counts()
demographic_data_counts

#Player Percentage by age
percent_players = (demographic_data_counts/number_players*100)
percent_players

#summary table
summary_table = pd.DataFrame({"Percentage of Players": round(percent_players,2),
                             "Total Count": demographic_data_counts})
summary_table.sort_index().head()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Establish bins for ages
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


player_data["Age Summary"] = pd.cut(player_data["Age"], bins, labels=group_names)

#Purchase count
purchasecount_data = player_data["Age Summary"].value_counts()

#Total Purchase
purchase_total = player_data.groupby(['Age Summary']).sum()

#Average Purchase
purchase_average = player_data.groupby(['Age Summary']).mean()

#summary table
summary_table = pd.DataFrame({"Purchase Count": purchasecount_data,
                              "Average Purchase Price": round(purchase_average['Price'],2),
                             "Total Purchase Value": purchase_total['Price']})
summary_table.head().sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting

```python
#Purchase Count
SN_data = player_data["SN"].value_counts()
SN_data

#Total Purchase
SN_total = player_data.groupby(['SN']).sum()
SN_total['Price']

#Average Purchase
SN_price = player_data.groupby(['SN']).mean()
SN_price['Price']

#Summary Table
summary_table = pd.DataFrame({"Purchase Count": SN_data,
                              "Average Purchase Price": round(SN_price['Price'],2),
                             "Total Purchase Value": SN_total['Price']})
summary_table.sort_values(["Total Purchase Value"], ascending=False).head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#lookup Items
item_data = purchase_data.loc[:,['Item ID','Item Name','Price']]

#Total Purchase#Purchase count
total_item = item_data.groupby(['Item ID','Item Name']).sum()

#Average Purchase
item_price= item_data.groupby(['Price'])

#Purchase count
count_item = item_data.groupby(['Item ID', 'Item Name']).count()

#Assort columns and rows
count_item.sort_values(["Price"], ascending=False).head(5)

#summary table
summary_table = pd.DataFrame({"Purchase Count": count_item['Price'],
                             "Average Purchase Price": total_item['Price']/count_item['Price'],
                            "Total Purchase Value": total_item['Price']})
summary_table.sort_values(["Purchase Count"], ascending=False).head(5)

```
