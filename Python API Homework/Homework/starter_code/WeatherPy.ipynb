```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import openweathermapy as ow
import pandas as pd
import numpy as np
import json
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# Testing API Call

# OpenWeatherMap API Key api_keys name of file second api_key refers to one we store it in
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?&APPID=" + api_key

testurl = url + "&q=birmingham"
test_data = requests.get(testurl).json()

print(test_data)
```

```python
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?&APPID=" + api_key 
#list creation
weather_data = []

#loop
for city in cities:
    cityurl = url + "&q=" + city
    
    time.sleep(5)
    
    print("processing record %s" %city)
    try:
        city_data = requests.get(cityurl).json()
        # Temperature (F) 
        citytemp = city_data["main"]["temp_max"]
        citycloudiness = city_data["clouds"]["all"]
        citylongitude = city_data["coord"]["lon"]
        citylatitude = city_data["coord"]["lat"]
        cityhumidity = city_data["main"]["humidity"]
        citycountry = city_data["sys"]["country"]
        citydate = city_data["dt"]
        citywindspeed = city_data["wind"]["speed"]
        
        print("found %s" %city)
     # appending city information into weather_data list
        weather_data.append({
            "City": city,
            "Cloudiness": citycloudiness,
            "Country": citycountry,
            "Date": citydate,
            "Humidity": cityhumidity,
            "Latitude": citylatitude,
            "Longitude": citylongitude,
            "Max Temp": citytemp,
            "Wind Speed": citywindspeed
            
            })
    
    except:
        print("city not found %s" %city)
        pass
    
    


```

```python
pd_weather_data = pd.DataFrame(weather_data)
pd_weather_data.head()
```

```python
# Bubble Scatter Plot
LatitudeVSMaxTemp = plt.scatter(x = pd_weather_data["Latitude"],y = pd_weather_data["Max Temp"] , marker="o", facecolors="red", edgecolors="black", alpha =0.5)
#labels
plt.title("City Latitude vs Max Temperature (08/15/2018)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

#grid
plt.grid()
```
